```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/자연휴양림출력화일.xlsx')
df.head()
```

```python
anMap = folium.Map(location=[37.4965,126.9573],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale=True)
anMap
```

```python
for n, row in df.iterrows():
  print(n, row['휴양림명'], row['위도'], row['경도'])
```

```python
for n, row in df.iterrows():
  mk = folium.Marker(location=[row['위도'], row['경도']], popup=folium.Popup(row['휴양림명'],max_width=200))
  mk.add_to(anMap)
anMap
```

```python
anMap2 = folium.Map(location=[37.4965, 126.9573],
                    zoom_start=10,
                    zoom_control=False,
                    control_scale=True)
anMap2
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap2)

for _, row in df.iterrows():
    folium.Marker(location=[row["위도"], row["경도"]],
                  popup=folium.Popup(row["휴양림명"], max_width=200)
                 ).add_to(mc)

anMap2
```
